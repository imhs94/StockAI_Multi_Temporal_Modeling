# 데이터 가져오기
## 기본 raw 데이터

```python
lst_stock_adjust = [] 
# 데이터 가져올 기간
start_date = '20130101'
end_date = '20230801'

# 보조지표 추가
for line in tqdm(kospi200): 
    code = line.strip() # KOSPI 200 중 상장일 2011년 이전 종목
    
    # 기본 주가 정보 불러오기
    stock_data = stock.get_market_ohlcv_by_date(start_date, end_date, code, adjusted=True)
    
    # 액면분할로 인한 거래량 보정을 위해 거래대금 불러오기
    stock_adjust = stock.get_market_ohlcv_by_date(start_date, end_date, code, adjusted=False) # 액면분할 보정된 거래대금 불러오기
    trading = stock_adjust['거래대금']

    # 시가총액 불러오기
    stock_cap = stock.get_market_cap(start_date, end_date, code)
    capitalization = stock_cap['시가총액']
    
    # 거래량 보정
    stock_data['거래량'] = stock_adjust['거래대금'] // stock_data['종가']
    
    # 두가지 데이터 합쳐주기
    stock_data = pd.concat([stock_data, trading, capitalization], axis=1)
    
    # 종목, 날짜 추가
    stock_data['code'] = code
    stock_data.index = stock_data.index.strftime('%Y-%m-%d') # 날짜형식 str형태로 바꿔주기 
    stock_data = stock_data.reset_index('날짜')
    lst_stock_adjust += stock_data.values.tolist()
```

# 보조지표 가져오기
## 보조지표 추가 함수 정의

```python
# 보조지표 추가
# SMA, EMA, WMA , KAMA, Ichimoku, Parabolic SAR(삭제하기로 한 것)
def calculate_technical_indicators(stock): 
    ma = [5,20,60,120]
    indicators = []
    for days in ma:
        indicator_name = 'MA' + str(days)
        indicators.append(indicator_name)
        stock[indicator_name] = stock['종가'].rolling(window=days).mean()
    
    H, L, C, V = stock['고가'], stock['저가'], stock['종가'], stock['거래량']

    # Volume -> Price
    stock['MFI'] = ta.volume.money_flow_index(
    high=H, low=L, close=C, volume=V, fillna=True) 
    stock['ADI'] = ta.volume.acc_dist_index(
    high=H, low=L, close=C, volume=V, fillna=True) 
    stock['OBV'] = ta.volume.on_balance_volume(close=C, volume=V, fillna=True)
    stock['CMF'] = ta.volume.chaikin_money_flow(
    high=H, low=L, close=C, volume=V, fillna=True) 
    stock['FI'] = ta.volume.force_index(close=C, volume=V, fillna=True)
    stock['EOM'] = ta.volume.ease_of_movement(
    high=H, low=L, volume=V, fillna=True)
    stock['EMV'] = ta.volume.ease_of_movement(
    high=H, low=L, volume=V, fillna=True) 
    stock['VPT'] = ta.volume.volume_price_trend(close=C, volume=V, fillna=True)
    stock['NVI'] = ta.volume.negative_volume_index(close=C, volume=V, fillna=True)
    stock['VMAP'] = ta.volume.volume_weighted_average_price(
    high=H, low=L, close=C, volume=V, fillna=True)
    
    # Volatility
    stock['ATR'] = ta.volatility.average_true_range(
    high=H, low=L, close=C, fillna=True)
    stock['BHB'] = ta.volatility.bollinger_hband(close=C, fillna=True)
    stock['BLB'] = ta.volatility.bollinger_lband(close=C, fillna=True)
    stock['KCH'] = ta.volatility.keltner_channel_hband(
    high=H, low=L, close=C, fillna=True)
    stock['KCL'] = ta.volatility.keltner_channel_lband(
    high=H, low=L, close=C, fillna=True)
    stock['KCM'] = ta.volatility.keltner_channel_mband(
    high=H, low=L, close=C, fillna=True)
    stock['DCH'] = ta.volatility.donchian_channel_hband(
    high=H, low=L, close=C, fillna=True)
    stock['DCL'] = ta.volatility.donchian_channel_lband(
    high=H, low=L, close=C, fillna=True)
    stock['DCM'] = ta.volatility.donchian_channel_mband(
    high=H, low=L, close=C, fillna=True)
    stock['UI'] = ta.volatility.ulcer_index(close=C, fillna=True)
    
    # Trend
    stock['SMA'] = ta.trend.sma_indicator(close=C, fillna=True)
    stock['EMA'] = ta.trend.ema_indicator(close=C, fillna=True)
    stock['WMA'] = ta.trend.wma_indicator(close=C, fillna=True)
    stock['MACD'] = ta.trend.macd(close=C, fillna=True)
    stock['ADX'] = ta.trend.adx(high=H, low=L, close=C, fillna=True)
    stock['-VI'] = ta.trend.vortex_indicator_neg(
    high=H, low=L, close=C, fillna=True)
    stock['+VI'] = ta.trend.vortex_indicator_pos(
    high=H, low=L, close=C, fillna=True)
    stock['TRIX'] = ta.trend.trix(close=C, fillna=True)
    stock['MI'] = ta.trend.mass_index(high=H, low=L, fillna=True)
    stock['CCI'] = ta.trend.cci(high=H, low=L, close=C, fillna=True)
    stock['DPO'] = ta.trend.dpo(close=C, fillna=True)
    stock['KST'] = ta.trend.kst(close=C, fillna=True)
    stock['Ichimoku'] = ta.trend.ichimoku_a(high=H, low=L, fillna=True)
    stock['STC'] = ta.trend.stc(close=C, fillna=True)
    
    # Momentum
    stock['RSI'] = ta.momentum.rsi(close=C, fillna=True)
    stock['SRSI'] = ta.momentum.stochrsi(close=C, fillna=True)
    stock['TSI'] = ta.momentum.tsi(close=C, fillna=True)
    stock['UO'] = ta.momentum.ultimate_oscillator(
    high=H, low=L, close=C, fillna=True)
    stock['SR'] = ta.momentum.stoch(close=C, high=H, low=L, fillna=True)
    stock['WR'] = ta.momentum.williams_r(high=H, low=L, close=C, fillna=True)
    stock['AO'] = ta.momentum.awesome_oscillator(high=H, low=L, fillna=True)
    stock['KAMA'] = ta.momentum.kama(close=C, fillna=True)
    stock['ROC'] = ta.momentum.roc(close=C, fillna=True)
    stock['PPO'] = ta.momentum.ppo(close=C, fillna=True)
    stock['PVO'] = ta.momentum.pvo(volume=V, fillna=True)

    return stock, indicators
```

## 일별 데이터 생성(보조지표 추가)

```python
# 결과 저장용 리스트
lst_code2date = []
lst_dayX = []
lst_dayY = []

# 일봉 보조지표 추가를 위해 기간 연장
start_date = '20110101' 
end_date = '20230801'

# 보조지표 추가
for line in tqdm(kospi200): 
    
    code = line.strip() # KOSPI 200 중 상장일 2011년 이전 종목
    
    # 주가 정보 불러오기
    stock_data = stock.get_market_ohlcv_by_date(start_date, end_date, code, adjusted=True) 
    
    # 액면분할 보정된 거래대금 불러오기
    stock_adjust = stock.get_market_ohlcv_by_date(start_date, end_date, code, adjusted=False) 
    trading = stock_adjust['거래대금']
    
    # 시가총액 불러오기
    stock_cap = stock.get_market_cap(start_date, end_date, code)
    capitalization = stock_cap['시가총액']
    
    # 거래량 보정
    stock_data['거래량'] = stock_adjust['거래대금'] // stock_data['종가']

    # 데이터 합쳐주기
    stock_data = pd.concat([stock_data, trading, capitalization], axis=1)
    
    # 시가, 고가, 저가가 0인 경우 해당 행 삭제 (거래정지로 인한 0)
    zero_mask = (stock_data[['시가', '고가', '저가']] == 0).any(axis=1)
    stock_data = stock_data[~zero_mask]
    
    # 보조지표 계산
    stock_with_indicators, indicators = calculate_technical_indicators(stock_data)

    # 일별 데이터에 55개?의 보조지표 추가(수정)
    stock_data = pd.concat([stock_data, stock_with_indicators[indicators]], axis=1)
    stock_data = stock_data.loc[:, ~stock_data.columns.duplicated()]
    
    # 다음날 종가 상승률(NEXT_CHANGE) 구하기
    stock_data['Next_Change'] = (stock_data['종가'].shift(-1) / stock_data['종가']) - 1

    stock_data.index = stock_data.index.strftime('%Y-%m-%d') # 날짜형식 str형태로 바꿔주기 
    stock_data = stock_data.reset_index('날짜')
    
    # 학습에 사용하지 않는 변수 삭제
    stock_data.drop(['등락률'], axis=1, inplace=True)
    stock_data.drop(columns=['거래량'], inplace=True) 
    
    lst_stock = stock_data.values.tolist() # 데이터프레임 형태에서 리스트 형태로 바꿔주기

    # 데이터 결과에 넣기
    for i, row in enumerate(lst_stock):

        lst_result = []

        date = row[0]
        Open, High, Low, Close, Trading, capitalization = row[1:7] 
        indicator = row[7:-1] # 보조지표
        lst_result.extend([Open, High, Low, Close, Trading, capitalization] + indicator) # 10일간의 주가 데이터 저장
            
        next_change = row[-1] # 다음날 종가 상승률
            
        lst_code2date.append([code, date]) # 거래대금 10억 이상인 종목과 날짜 저장
        lst_dayX.append(lst_result) # 10일간의 주가 데이터 저장 
        lst_dayY.append([next_change]) # next_change 값과 target값 저장
```

```python
# 컬럼값 생성
indicator = ['Open', 'High', 'Low', 'Close','Trading', 'capitalization', 'MA5', 'MA20', 'MA60', 'MA120',
                   'MFI', 'ADI', 'OBV', 'CMF', 'FI', 'EOM', 'EMV', 'VPT', 'NVI', 'VMAP', 'ATR',
                   'BHB', 'BLB', 'KCH', 'KCL', 'KCM', 'DCH', 'DCL', 'DCM', 'SMA', 'EMA', 'WMA',
                   'Ichimoku', 'KAMA', 'MACD', 'UI', 'ADX', '-VI', '+VI', 'TRIX',
                   'MI', 'CCI', 'DPO', 'KST', 'STC', 'RSI', 'SRSI', 'TSI', 'UO', 'SR', 'WR', 'AO',
                   'ROC', 'PPO', 'PVO']
```

```python
day_code2date = pd.DataFrame(lst_code2date, columns = ['code','date']) # 일별 종목, 날짜 저장
day_X = pd.DataFrame(lst_dayX, columns = indicator) # 일별 주가 정보 저장
day_Y = pd.DataFrame(lst_dayY, columns = ['next_change']) # 다음날 종가 상승률, label 값 저장
df_day = pd.concat([day_code2date,day_X,day_Y], axis = 1) # 보조지표 추가된 일별 데이터 프레임생성
df_day
```

```python
# 스케일링이 필요한 보조지표 
col_scaling = ['Open', 'High', 'Low', 'Close', 'Trading', 'capitalization', 'MA5', 'MA20', 'MA60', 'MA120', 
           'VMAP', 'BHB', 'BLB', 'KCH', 'KCL', 'KCM', 'DCH', 'DCL', 'DCM',
           'SMA', 'EMA', 'WMA', 'Ichimoku', 'KAMA','MACD']

# 스케일링이 필요하지 않은 보조지표
col_no_scaling = ['MFI', 'ADI', 'OBV', 'CMF', 'FI','EOM', 'EMV', 'VPT', 'NVI', 'ATR', 'UI', 'ADX',
       '-VI', '+VI', 'TRIX', 'MI', 'CCI', 'DPO', 'KST', 'STC', 'RSI', 'SRSI',
       'TSI', 'UO', 'SR', 'WR', 'AO', 'ROC', 'PPO', 'PVO', 'next_change']
```

```python
df_day['BeforeClose'] = df_day['Close'].shift(1) # 전날 종가 컬럼 추가
df_day
```

```python
# code, date 따로 저장
day_code = df_day['code']
day_date = df_day['date']
day_code2date = pd.DataFrame({'code': day_code, 'date': day_date})
```

```python
# 전날 종가로 스케일링
day_scale_stock = df_day[col_scaling].apply(lambda x: x/df_day['BeforeClose'])
# 스케일링이 필요없는 지표는 따로 저장
day_no_scale_stock = df_day[col_no_scaling]
```

```python
# 로그화이후 데이터 합치기
scale_day = pd.concat([day_code2date,day_log_stock,day_no_log_stock], axis = 1)
scale_day
```

## 주별 데이터

```python
# 2011년 이후의 일별 데이터 불러오기
df_day = pd.read_csv('../data/df_day.csv')
df_day
```

```python
# 일별 데이터로부터 주봉 데이터를 생성하는 함수
def WD(df_dy):
    df_day['date'] = pd.to_datetime(df_day['date'])
    weekly_data = pd.DataFrame(columns=df_day.columns)

    for code, code_data in df_day.groupby('code'): # 코드를 기준으로 그룹화
        code_weekly_data = code_data.copy()

        # 결과 담아줄 변수 생성
        weekly_open = None
        weekly_high = None
        weekly_low = None
        weekly_dollar_volume = 0
        last_weekday = None
        
        for i, row in code_data.iterrows():
            date = row['date']
            current_weekday = date.weekday()
            
            if weekly_open is None: # 새로운 주로 시작 되었는지 판단 -> weekly_open이 초괴화되었다면 새로운 주의 시작을 뜻함
                weekly_open = row['Open']
                weekly_high = row['High']
                weekly_low = row['Low']
                weekly_dollar_volume = row['Trading']
            elif current_weekday <= last_weekday:  # 시작 이후 데이터 채우기 -> 현재 요일이 지난 요일보다 작거나 같은 경우 새로운 주 시작
                weekly_open = row['Open']
                weekly_high = row['High']
                weekly_low = row['Low']
                weekly_dollar_volume = row['Trading']
            else:                                  # 현재 요일이 이전 요일보다 큰 경우 사실상 예외 값
                weekly_high = max(weekly_high, row['High'])
                weekly_low = min(weekly_low, row['Low'])
                weekly_dollar_volume += row['Trading']

            weekly_close = row['Close']  # 종가는 항상 마지막 값으로 채워주기

            # 데이터 순서대로 저장
            code_weekly_data.loc[i, 'Open'] = weekly_open
            code_weekly_data.loc[i, 'High'] = weekly_high
            code_weekly_data.loc[i, 'Low'] = weekly_low
            code_weekly_data.loc[i, 'Close'] = weekly_close
            code_weekly_data.loc[i, 'Trading'] = weekly_dollar_volume

            last_weekday = current_weekday

        # 각 종목 코드의 주봉 데이터를 합침
        weekly_data = pd.concat([weekly_data, code_weekly_data])

    return weekly_data
```

```python
# 일별로 쌓이는 주봉 데이터를 생성
week_data = WD(df_day)
print("일별로 쌓이는 주봉 데이터:\n", week_data)
```

```python
# 종목코드별로 그룹화
week_grouped = week_data.groupby('code')

# 종목코드별로 이동평균선을 계산하는 함수를 정의합니다.
def calc_ma(x):
    x['W_MA5'] = x['weekly_Close'].rolling(window=25).mean()
    x['W_MA10'] = x['weekly_Close'].rolling(window=50).mean()
    x['W_MA20'] = x['weekly_Close'].rolling(window=100).mean()
    return x

# 종목코드별로 계산 함수를 적용하여 이동평균선을 계산합니다.
week_data = week_grouped.apply(calc_ma)
week_data
```

```python
# 일별 데이터로부터 월별 데이터를 생성하는 함수
def MD(df_day):
    df_day['date'] = pd.to_datetime(df_day['date'])
    monthly_data = pd.DataFrame(columns=df_day.columns)

    for code, code_data in df_day.groupby('code'): # 코드를 기준으로 그룹화
        code_monthly_data = code_data.copy()

        monthly_open = None
        monthly_high = None
        monthly_low = None
        monthly_volume = 0
        last_month = None

        for i, row in code_data.iterrows():
            date = row['date']
            current_month = date.month

            if last_month is None:
                last_month = current_month

            if monthly_open is None: # 새로운 주로 시작 되었는지 판단 -> weekly_open이 초괴화되었다면 새로운 주의 시작을 뜻함
                monthly_open = row['Open']
                monthly_high = row['High']
                monthly_low = row['Low']
                monthly_dollar_volume = row['Trading']
            elif current_month != last_month:  # 시작 이후 데이터 채우기 -> 현재 요일이 지난 요일보다 작거나 같은 경우 새로운 주 시작
                monthly_open = row['Open']
                monthly_high = row['High']
                monthly_low = row['Low']
                monthly_dollar_volume = row['Trading']
            else:                                  # 현재 요일이 이전 요일보다 큰 경우 사실상 예외 값
                monthly_high = max(monthly_high, row['High'])
                monthly_low = min(monthly_low, row['Low'])
                monthly_dollar_volume += row['Trading']

            monthly_close = row['Close'] # 종가는 항상 마지막 값으로 채워주기

            # 데이터 순서대로 저장
            code_monthly_data.loc[i, 'Open'] = monthly_open
            code_monthly_data.loc[i, 'High'] = monthly_high
            code_monthly_data.loc[i, 'Low'] = monthly_low
            code_monthly_data.loc[i, 'Close'] = monthly_close
            code_monthly_data.loc[i, 'Trading'] = monthly_dollar_volume

            last_month = current_month

        # 각 종목 코드의 월봉 데이터를 합치기
        monthly_data = pd.concat([monthly_data, code_monthly_data])

    return monthly_data
```

## 10일치 데이터 생성

```python
# 전체 데이터 불러오기
data = pd.read_csv('../data/scale_all_data_modify.csv')
data
```

```python
# 10일치 데이터로 바꾸기
data['date'] = pd.to_datetime(data['date'])
# 종목 리스트 추출
stock_list = data['code'].unique()

# 결과를 저장할 리스트 생성
data_list = []

# 각 종목별로 오늘을 포함한 이전 9일의 데이터 추출하여 새로운 DataFrame에 추가
for stock in stock_list:
    # 해당 종목의 주가 데이터 추출
    stock_data = data[data['code'] == stock]
    
    # 오늘을 기준으로 이전 9일의 데이터 추출 (모든 컬럼)
    window_size = 10
    
    # 유효한 행 범위 계산 (첫 9행 무시)
    valid_range = len(stock_data) - window_size + 1
    
    if valid_range > 0:
        for i in range(window_size-1, len(stock_data)):
            # 윈도우 크기 범위 내에서 해당 행과 이전 9개 행 가져오기 (모든 컬럼)
            Ddays_date = stock_data.iloc[i-window_size+1:i+1][::-1]   # [::-1] is added to reverse the order

            # lst_result에 최근 일별 모든 정보를 한 행에 담아줌 
            lst_result = []
            for row2 in Ddays_date.values:
                lst_result += list(row2)

            data_list.append(lst_result)

# 리스트를 DataFrame으로 변환하여 결과 출력
new_df = pd.DataFrame(data_list)

column_names=[]
for i in range(window_size):
    for column_name in data.columns:
        column_names.append(f'{column_name}_D-{i}')   # changed from 'window_size-i-1' to 'i'

new_df.columns=column_names

new_df
```

```python
# 삭제될 컬럼의 리스트 생성 -> 중복되는 code, date, next_change 삭제
columns_to_drop = []

for i in range(1, 10): # D-1부터 D-9까지 반복 -> D-0인 현재만 필요함
    columns_to_drop.extend([f"code_D-{i}", f"date_D-{i}", f"next_change_D-{i}"])

# 삭제할 컬럼이 있는지 확인하고 삭제
filtered_columns = [col for col in new_df.columns if col in columns_to_drop]
new_df = new_df.drop(filtered_columns, axis=1)

new_df
```

```python
# 모델링을 위해 target값인 next_change 맨 뒤로 보내기
next_change = new_df['next_change_D-0']
new_df.drop(['next_change_D-0'], axis=1, inplace=True)
```

```python
# 열 이름 변경
next_change = next_change.to_frame('next_change')
next_change
```

```python
# next_change 맨 뒤로 보내주기 및 code, date로 명 변경
scale_all_data_10 = pd.concat([new_df, next_change], axis=1)
scale_all_data_10 = scale_all_data_10.rename(columns={'code_D-0': 'code', 'date_D-0': 'date'})
scale_all_data_10
```
